# Analyzing Indian Political Parties' Tweets with Machine Learning

## Introduction

In this project, we'll delve into the world of Indian politics by analyzing the tweets of three major political parties — Bharatiya Janata Party (BJP), Indian National Congress (Congress), and Aam Aadmi Party (AAP). Leveraging the power of Natural Language Processing (NLP) and machine learning, we aim to understand the political discourse and underlying sentiments that are represented in these tweets.

## Workflow Overview

### 1. Environment Setup

To kick things off, we ensure all necessary Python libraries and packages are installed and imported. This includes data manipulation library `pandas`, and various modules from the `transformers` package, which offers state-of-the-art machine learning models for NLP.

### 2. Data Acquisition

We then load the dataset, which is a compilation of tweets from the aforementioned political parties. By consolidating these tweets into a single pandas DataFrame, we establish a comprehensive base for our forthcoming analysis.

### 3. Data Sampling

Given the potentially large volume of tweets, we extract a random sample of 10,000 tweets. This helps manage computational resources while still providing a representative subset for analysis.

### 4. Translation and Classification

To cater to our multilingual dataset, we've defined a `Translator` class to convert all tweets into English. Meanwhile, the `PoliticalBiasClassifier` class is used to categorize the political leaning of each tweet into 'Left', 'Center', or 'Right'.

### 5. Data Processing

The actual translation and classification of tweets are executed in this phase. Each tweet in our sample dataset is translated and subsequently classified for political bias. The outcomes are stored in lists for later use.

### 6. DataFrame Update

Post-processing, our original DataFrame is updated with the translated text and corresponding political bias classifications. This enriched dataset is saved as a CSV file for possible future use and analysis.

## About the Dataset

Our dataset is a dynamic one, comprising daily updated tweets from BJP, Congress, and AAP. These tweets span various topics, providing a robust insight into the diverse landscape of political discourse in India.

Each row in the dataset signifies a tweet and includes details like:

- Datetime: When the tweet was posted.
- Tweet Id: The unique identifier of the tweet.
- Text: The text content of the tweet.
- Username: The account that posted the tweet.
- likeCount: The number of likes received by the tweet.

```python
import pandas as pd
```

```python
INC = pd.read_csv('/kaggle/input/indian-political-party-tweets-daily-updated/INCIndia.csv')
AAP = pd.read_csv('/kaggle/input/indian-political-party-tweets-daily-updated/AamAadmiParty.csv')
BJP = pd.read_csv('/kaggle/input/indian-political-party-tweets-daily-updated/BJP4India.csv')
df = pd.concat([AAP, BJP, INC])
```

```python
sample_df = df.sample(10000,random_state = 42)
```

```python
sample_df.info()
```

```python
import pandas as pd
from transformers import MarianMTModel, MarianTokenizer, AutoTokenizer, AutoModelForSequenceClassification
import torch
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class Translator:
    def __init__(self, model_name):
        self.model = MarianMTModel.from_pretrained(model_name).to(device)
        self.tokenizer = MarianTokenizer.from_pretrained(model_name)

    def translate(self, sentence):
        tokenized_text = self.tokenizer.encode(sentence, return_tensors="pt").to(device)
        translated = self.model.generate(tokenized_text)
        return self.tokenizer.decode(translated[0], skip_special_tokens=True)

class PoliticalBiasClassifier:
    def __init__(self, model_name):
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModelForSequenceClassification.from_pretrained(model_name).to(device)

    def classify(self, text):
        inputs = self.tokenizer.encode_plus(text, truncation=True, padding='max_length', max_length=128, return_tensors='pt').to(device)
        outputs = self.model(**inputs)
        bias_map = {0: "Left", 1: "Center", 2: "Right"}
        scores = torch.nn.functional.softmax(outputs.logits, dim=1).tolist()[0]
        bias_scores = {bias_map[i]: scores[i] for i in range(3)}
        return bias_scores

translator = Translator('Helsinki-NLP/opus-mt-mul-en')
classifier = PoliticalBiasClassifier('bucketresearch/politicalBiasBERT')

# Assume you have a DataFrame named sample_df
# Read the DataFrame for example
# sample_df = pd.read_csv('/kaggle/input/your-data.csv')

def process_row(row):
    translated_text = translator.translate(row['Text'])
    bias_score = classifier.classify(translated_text)
    return pd.Series([translated_text, bias_score['Left'], bias_score['Center'], bias_score['Right']])

tqdm.pandas()
sample_df[['Translated_Text', 'Left', 'Center', 'Right']] = sample_df.progress_apply(process_row, axis=1)
sample_df.to_csv('/kaggle/working/Sample_Data_Political_Sentiment(10k).csv', index=False)
```

# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate political bias scores
bias_score_agg = sample_df[['Left', 'Center', 'Right']].mean()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=bias_score_agg.index, y=bias_score_agg.values, palette='viridis')
plt.title('Average Political Bias Scores')
plt.ylabel('Average Score')
plt.xlabel('Political Bias')
plt.show()
```

```python
# Create new dataframe to store summed likeCount for each political bias
df_likes = pd.DataFrame()

# Sum likeCount for each political bias
df_likes['Left'] = [sample_df.loc[sample_df['Left'].idxmax(), 'likeCount']]
df_likes['Center'] = [sample_df.loc[sample_df['Center'].idxmax(), 'likeCount']]
df_likes['Right'] = [sample_df.loc[sample_df['Right'].idxmax(), 'likeCount']]

# Convert dataframe from wide format to long format for visualization
df_likes_melt = df_likes.melt(var_name='Political Bias', value_name='likeCount')

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Political Bias', y='likeCount', data=df_likes_melt, palette='viridis')
plt.title('Total Likes for Each Political Bias')
plt.ylabel('Likes')
plt.xlabel('Political Bias')
plt.show()
```
