```python
# Import the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/kaggle/input/top-250-imdb-series-episode-ratings/imdb_top_250_series_episode_ratings.csv')
df = df.drop('Unnamed: 0',axis = 1)
```

```python
import plotly.express as px

df_Q1 = df[['Title','Rating']].groupby('Title').mean().round(2).sort_values(by='Rating', ascending=False).nlargest(10,'Rating').reset_index()

fig = px.bar(df_Q1, x='Title', y='Rating',
              color='Title',
             labels={'Title':'TV Series by their Mean Episode Rating'}, height=400,text_auto=True)
fig.show()
```

**Chernobyl is the top TV series by their mean episode rating**

```python
import plotly.express as px

df_Q2 = df[['Title','Rating']].groupby('Title').median().round(2).sort_values(by='Rating', ascending=False).nlargest(10,'Rating').reset_index()

fig = px.bar(df_Q2, x='Title', y='Rating',
              color='Title',
             labels={'Title':'TV Series by their Median Episode Rating'}, height=400,text_auto=True)
fig.show()
```

**Chernobyl is the top TV series by their median episode rating**

```python
df_y = df.groupby(['Season','Title']).mean().drop('Episode',axis = 1).reset_index().sort_values(by='Rating', ascending=False).nlargest(10,'Rating').reset_index()
fig = px.bar(df_y, x="Title", y="Rating",
             color='Season',
             labels={'Title':'TV Season by their Mean Episode Rating'}, height=400,text_auto=True)
fig.show()
```

**Chernobyl season 1 is the top TV season by their mean episode rating**

```python
df_y = df.groupby(['Season','Title']).median().drop('Episode',axis = 1).reset_index().sort_values(by='Rating', ascending=False).nlargest(10,'Rating').reset_index()
fig = px.bar(df_y, x="Title", y="Rating",
             color='Season',
             labels={'Title':'TV Season by their Median Episode Rating'}, height=400,text_auto=True)
fig.show()
```

**Chernobyl season 1 is the top TV season by their median episode rating**

```python
df_b = df.groupby(['Season','Title']).max().drop('Code',axis = 1).sort_values(by='Rating', ascending=False).nlargest(10,'Rating').reset_index()
fig = px.bar(df_b, x="Title", y="Rating",
            color='Season',
           labels={'Title':'The most loved TV Series starting (with the highest rating in the last episode)'}, height=400,text_auto=True)
fig.show()
```

**Breaking Bad season 5 is the highest rated ending episode with rating 10**

```python
df_a = df.query("Episode == 1")[['Episode','Rating','Title','Season']].reset_index().drop('index',axis = 1).sort_values(by='Rating', ascending=False).nlargest(10,'Rating').reset_index()
fig = px.bar(df_a, x="Title", y="Rating",
            color='Season',
           labels={'Title':'The most loved TV Series starting (with the highest rating in the first episode)'}, height=400,text_auto=True)
fig.show()
```

**House M.D. season 6 is the highest rated opening episode with rating 9.7**

```python
#What are the TV Series that continuously keep improving (with the highest slope in their rating trend)?
samp_2 = pd.DataFrame(columns = ['Season','Title','Rating','Episode'])
for i in list(df.Title.unique()):
    samp_1 = df[['Season','Title','Rating','Episode']].groupby('Title').get_group(i)
    samp_1['Percentage Change'] = samp_1['Rating'].pct_change()
    samp_2 = samp_2.append(samp_1, ignore_index=True)
samp_2 = samp_2.dropna()
```

```python
import plotly.express as px

fig = px.line(samp_2, x="Episode", y="Percentage Change", color='Title', animation_frame="Season",height = 1300)
fig.update_yaxes(showgrid=False),
fig.update_traces(hovertemplate=None)
fig.update_layout(margin=dict(t=70, b=0, l=70, r=40),
                        hovermode="x unified",
                        xaxis_tickangle=360,
                        xaxis_title=' ', yaxis_title=" ",
                        plot_bgcolor='#2d3035', paper_bgcolor='#2d3035',
                        title_font=dict(size=25, color='#a5a7ab', family="Lato, sans-serif"),
                        font=dict(color='#8a8d93'),
                        legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1))
fig.show()
```

```python
samp_3 = samp_2.sort_values(by='Percentage Change', ascending=False).nlargest(20,'Percentage Change').reset_index()
samp_3.drop('index',axis = 1)
```

**Table stating the Top 20 Percentage Changes with the highest positive slope change for Top Gear, Star Trek: The Next Generation and Hunter x Hunter are the top 3.**

```python
df_epi = df[['Title','Rating']].groupby('Title').mean()
df_epi.rename(columns = {'Rating':'Mean Episode Rating'}, inplace = True)
df_epi = df_epi.reset_index()
df_epi
```

```python
df_global = df_global[['Title','Rating']]
df_global.rename(columns = {'Rating':'TV Series Rating'}, inplace = True)
df_global
```

```python
df_diff = pd.merge(df_epi, df_global, on='Title', how='inner')
df_diff['Gap'] = df_diff['Mean Episode Rating'] - df_diff['TV Series Rating']
#Finding the absoulte value to get the maximum difference
df_diff['Gap'] = df_diff['Gap'].abs()
df_diff = df_diff.sort_values(by='Gap', ascending=False).nlargest(20,'Gap').reset_index().round(2)
fig = px.bar(df_diff, x="Title", y="Gap",
            color='Title',
           labels={'Title':'Top ten Gap between Mean Episode Rating and TV Series Rating'}, height=400,text_auto=True)
fig.show()
```

**Mahabharat has the highest gap Mean Episode Rating and TV Series Rating**
