# Load the Dataset

```python
import pandas as pd

# data with 'ISO-8859-1' encoding
youtube_data = pd.read_csv('/kaggle/input/global-youtube-statistics-2023/Global YouTube Statistics.csv', encoding='ISO-8859-1')

# Display the first few rows of the dataframe
youtube_data.head()
```

# Top 10 YouTube channels by subscriber count, video views and correlation between different metrics

```python
# Top 10 YouTube channels by subscriber count
top_channels_by_subscribers = youtube_data.nlargest(10, 'subscribers')[['Youtuber', 'subscribers']]

# Top 10 YouTube channels by video views
top_channels_by_views = youtube_data.nlargest(10, 'video views')[['Youtuber', 'video views']]

# Correlation between different metrics
correlation_metrics = youtube_data[['subscribers', 'video views', 'uploads']].corr()

top_channels_by_subscribers, top_channels_by_views, correlation_metrics
```

1. **Top 10 Channels by Subscribers**
   - The most subscribed YouTube channel is "T-Series" with 245 million subscribers, followed by "YouTube Movies" with 170 million and "MrBeast" with 166 million subscribers.

2. **Top 10 Channels by Video Views**
   - The YouTube channel with the most video views is "T-Series" with 228 billion views. "Cocomelon - Nursery Rhymes" and "SET India" follow closely, with 164 billion and 148 billion views, respectively.

3. **Correlation between Subscribers, Video Views, and Uploads**
   - There is a strong positive correlation (0.75) between the number of subscribers and the number of video views, indicating that channels with more subscribers tend to have more views.
   - The number of uploads has a low positive correlation with both the number of subscribers (0.07) and the number of video views (0.16). This suggests that simply uploading more videos does not necessarily lead to more subscribers or views.

This analysis provides an overview of the types of content produced by the YouTube channels in this dataset and the quantity of content they typically produce.

# Map View Countries and its Subscribers

```python
import folium
import math


# Initialize the map and set the initial location to the world center
world_map = folium.Map(location=[0, 0], zoom_start=2)

# Create a DataFrame that contains the top channel by subscribers in each country
top_channel_by_country = youtube_data.loc[youtube_data.groupby('Country')['subscribers'].idxmax()]

# Filter the DataFrame to only include rows with valid latitude and longitude values
top_channel_by_country = top_channel_by_country.dropna(subset=['Latitude', 'Longitude'])

# Add a marker for each top YouTube channel in each country
for idx, row in top_channel_by_country.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']],
                  popup=f"{row['Youtuber']}: {row['subscribers']} subscribers",
                  icon=folium.Icon(color='blue')).add_to(world_map)

# Display the map
world_map
```

# Channel Age Analysis, Channel Category Analysis, Upload Frequency Analysis and Country-specific Analysis

```python
from datetime import datetime

# Channel Age Analysis
# Compute the age of each channel in years
current_year = datetime.now().year
youtube_data['age'] = current_year - youtube_data['created_year']
channel_age_correlation = youtube_data[['age', 'subscribers', 'video views', 'highest_monthly_earnings', 'highest_yearly_earnings']].corr()

# Channel Category Analysis
# Compute the mean performance metrics for each category
category_analysis = youtube_data.groupby('category').agg({'subscribers': 'mean', 'video views': 'mean', 'highest_monthly_earnings': 'mean', 'highest_yearly_earnings': 'mean'})

# Upload Frequency Analysis
upload_frequency_correlation = youtube_data[['uploads', 'subscribers', 'video views', 'highest_monthly_earnings', 'highest_yearly_earnings']].corr()

# Country-specific Analysis
# Let's take the United States as an example
us_data = youtube_data[youtube_data['Country'] == 'United States']
us_top_channels = us_data.nlargest(10, 'subscribers')[['Youtuber', 'subscribers']]
us_popular_categories = us_data['category'].value_counts()

channel_age_correlation, category_analysis, upload_frequency_correlation, us_top_channels, us_popular_categories
```

1. **Correlation between Channel Age, Subscribers, and Video Views**
   - The correlation between channel age and subscribers is positive but weak (0.142), suggesting that older channels tend to have more subscribers, but the correlation is not strong.
   - The correlation between channel age and video views is also positive but weak (0.127), indicating that older channels tend to have more video views, but again, the correlation is not strong.
   - There is a negative correlation between channel age and both highest monthly earnings (-0.072) and highest yearly earnings (-0.072), suggesting that older channels do not necessarily earn more.

2. **Average Performance by Category**
   - The category with the highest average number of subscribers is "Shows" with 41,615,380 subscribers. 
   - The category with the highest average number of video views is "Shows" with 34,323,600,000 views.
   - The category with the highest average monthly earnings is "Shows" with $2,037,662.
   - The category with the highest average yearly earnings is also "Shows" with $24,384,620.

3. **Correlation between Uploads, Subscribers, Video Views, and Earnings**
   - The correlation between the number of uploads and the number of subscribers is positive but weak (0.077), suggesting that channels with more uploads tend to have more subscribers, but the correlation is not strong.
   - The correlation between the number of uploads and the number of video views is positive (0.166), suggesting that channels with more uploads tend to have more video views.

4. **Top 10 Channels Excluding "T-Series" and "YouTube Movies"**
   - The channel with the highest number of subscribers excluding "T-Series" and "YouTube Movies" is "MrBeast" with 166,000,000 subscribers.
   - The top 10 list also includes channels such as "Cocomelon - Nursery Rhymes", "Kids Diana Show", and "Vlad and Niki".

5. **Distribution of Categories Among the Top 200 Channels**
   - The most common category among the top 200 channels is "Music" (73 channels), followed by "Entertainment" (68 channels) and "People & Blogs" (38 channels).
   - Less represented categories in the top 200 include "Science & Technology", "Sports", "News & Politics", "Pets & Animals", and "Nonprofits & Activism".

# Education Analysis, Urbanization Analysis

```python
# Education Analysis
education_correlation = youtube_data[['Gross tertiary education enrollment (%)', 'subscribers', 'video views', 'highest_monthly_earnings', 'highest_yearly_earnings']].corr()

# Urbanization Analysis
urbanization_correlation = youtube_data[['Urban_population', 'subscribers', 'video views', 'highest_monthly_earnings', 'highest_yearly_earnings']].corr()

education_correlation, urbanization_correlation
```

1. **Correlation between Gross Tertiary Education Enrollment and Channel Metrics**
   - The correlation between gross tertiary education enrollment and subscribers is negative and weak (-0.007), suggesting that there is almost no relationship between the two.
   - The correlation between gross tertiary education enrollment and video views is also negative and weak (-0.015).
   - The correlation between gross tertiary education enrollment and both highest monthly earnings (-0.062) and highest yearly earnings (-0.062) is negative, suggesting that higher levels of tertiary education enrollment do not necessarily correlate with higher YouTube earnings.

2. **Correlation between Urban Population and Channel Metrics**
   - The correlation between urban population and subscribers is positive but weak (0.084), suggesting that channels from countries with larger urban populations do not necessarily have more subscribers.
   - The correlation between urban population and video views is also positive but weak (0.077), suggesting that channels from countries with larger urban populations do not necessarily have more video views.
   - The correlation between urban population and both highest monthly earnings (0.081) and highest yearly earnings (0.081) is positive but weak, suggesting that channels from countries with larger urban populations do not necessarily earn more.

These points indicate that factors such as the level of tertiary education enrollment and the size of the urban population in a country do not have a strong impact on the success of YouTube channels in terms of subscribers, video views, or earnings.

# Performance by Creation Date

```python
# Compute the mean performance metrics for each creation year
performance_by_year = youtube_data.groupby('created_year').agg({'subscribers': 'mean', 'video views': 'mean', 'highest_monthly_earnings': 'mean', 'highest_yearly_earnings': 'mean'})

# Compute the mean performance metrics for each creation month
performance_by_month = youtube_data.groupby('created_month').agg({'subscribers': 'mean', 'video views': 'mean', 'highest_monthly_earnings': 'mean', 'highest_yearly_earnings': 'mean'})

# 4. Correlation between Population and Performance
population_correlation = youtube_data[['Population', 'subscribers', 'video views', 'highest_monthly_earnings', 'highest_yearly_earnings']].corr()

performance_by_year, performance_by_month, population_correlation
```

1. **Correlation between Channel Creation Year and Channel Metrics**
   - Channels created in earlier years tend to have higher subscriber counts and video views, with channels created in 2006, 2007, and 2008 showing the highest subscriber counts and video views on average. 
   - Interestingly, channels created in 2020 and 2021 have the highest average monthly and yearly earnings, suggesting that newer channels have found ways to monetize their content more effectively.

2. **Correlation between Channel Creation Month and Channel Metrics**
   - The month in which a channel was created does not seem to have a strong impact on subscriber count or video views, as these numbers are relatively consistent across all months. 
   - However, channels created in September, March, and August have the highest average monthly and yearly earnings, suggesting that these months may be more favorable for launching a new channel.

3. **Correlation between Population and Channel Metrics**
   - The correlation between population and subscribers is positive but weak (0.082), suggesting that channels from countries with larger populations do not necessarily have more subscribers.
   - The correlation between population and video views is also positive but weak (0.080), suggesting that channels from countries with larger populations do not necessarily have more video views.
   - The correlation between population and both highest monthly earnings (0.105) and highest yearly earnings (0.105) is positive but weak, suggesting that channels from countries with larger populations do not necessarily earn more.

These points indicate that factors such as the year and month of channel creation and the population size of a country do not have a strong impact on the success of YouTube channels in terms of subscribers, video views, or earnings. However, newer channels do appear to have higher earning potentials.
